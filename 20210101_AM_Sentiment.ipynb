# Sentiment Scores

Thank you to Stéfan Sinclair, Zoë Wilkinson Saldaña and Sunyam Bagga for code examples and inspiration.

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sentimentAnalyzer = SentimentIntensityAnalyzer() #Initialize sentiment scorer
```

```python
import nltk
from nltk import word_tokenize
sentenceTokenizer = nltk.data.load('tokenizers/punkt/english.pickle') #Load tokenizer
```

Deleted the following files from the dataset because on close inspection they appeared to be more in the form of journal entries than letters: S9908-D039.txt, S9908-D038.txt, S9908-D037.txt, S9908-D036.txt, S9908-D035.txt, S9908-D034.txt, S9908-D033.txt, S9908-D032.txt, S9908-D031.txt, S9908-D030.txt

```python
from nltk.corpus import PlaintextCorpusReader
lettersCorpus = PlaintextCorpusReader('letters', '.*txt') #Define corpus
len(lettersCorpus.fileids()) #How many files
```

```python
import pandas as pd
df = pd.DataFrame(columns = ['Sentence','fileid','Sequence','Sentiment']) # Create a new dataframe to hold sentences, letter id and compound scores
df.head()
```

```python
import re

for fileid in lettersCorpus.fileids(): # For each file in the corpus
    text = lettersCorpus.raw(fileid) # Place the string into the object "text"
    text = re.sub(r"(\. ){2,}","", text) # Remove ellipses
    sentences = sentenceTokenizer.tokenize(text) # Place sentences into a list called "sentences"
    sequence = 0 # Create a counter to keep track of sentence order
    for sentence in sentences: # For each sentence in this list  
        sequence +=1 # Counter updater
        sentence = re.sub(r"(\n+)"," ", sentence) # Replace blank lines with a single space
        sentence = re.sub(r"(&dot)","", sentence) # Remove this expression (a dot)
        sentence = re.sub(r"(— )","", sentence) # Remove dashes
        sentence = sentence.lower() # Make everything lowercase (because some docs in all caps)
        scores = sentimentAnalyzer.polarity_scores(sentence) # Calculate sentiment scores
        new_row = {'Sentence':sentence, 'fileid':fileid, 'Sequence':sequence,'Sentiment':scores['compound']} # Place the sentence, letter id and compound score for sentence into a dictionary called "new_row" in which the key is the variable and the value is the data
        df = df.append(new_row, ignore_index=True) # Populate a new row in the dataframe with the dictionary info
    print(fileid) # Show progress
```

```python
# Add a column "Sentences" showing total number of sentences in letter
df['Sentences'] = df.groupby('fileid')['Sequence'].transform('max')
df
```

```python
# Add a column "Position" showing the location of the sentence relative to the whole 
# That is, how far through the letter does the sentence appear?
df['Position'] = df['Sequence'] / df['Sentences']
df
```

```python
# Add a column "Last" showing whether the sentence is the last in a letter
df['Last'] = "False"
df.head()
```

```python
import numpy as np
df['Last'] = np.where((df.Position == 1),'True',df.Last)
df
```

```python
# Add a column "SentimentLTR" showing average sentiment for letter
df['SentimentLTR'] = df.groupby('fileid')['Sentiment'].transform('mean')
df
```

```python
df.to_csv('20210101_AM_sentiment.csv', index = True)
```
